# Using POYO

This notebook will walk you through three concepts:
1. Loading a dataset using our `Dataset` object
2. Loading the POYO model and applying pretrained weights
3. Performing inference and measuring accuracy

```python
import torch
import matplotlib.pyplot as plt
import random

import omegaconf                 # for loading model configuration from a file
from torchmetrics import R2Score # for measuring accuracy

from kirby.data import Dataset, Collate, build_vocab
from kirby.data.stitcher import stitched_prediction
from kirby.models import PerceiverNM
import kirby.taxonomy
```

## Loading the dataset
Let's first load a single session recording from [Perich-Miller's dataset](https://dandiarchive.org/dandiset/000688). 

```python
dataset = Dataset(
    root="/kirby/processed",
    split="test", # "train"/"valid"/"test"
    include=[{
        "selection": {
            "dandiset": "perich_miller",
            "sortset": "chewie_20131003",
        },
    }],
)

# These are needed by the model to build its unit and session embedding tables
unit_vocab = build_vocab(dataset.unit_names)
session_names = dataset.session_names
```

Now let's pick a random sample and plot its spike train

```python
sample_idx = random.randrange(len(dataset))
data_sample = dataset[sample_idx]

num_units = len(data_sample.units.unit_name)
spike_times = data_sample.spikes.timestamps
unit_ids = data_sample.spikes.unit_index

# Separate spike times for each unit
spikes_per_unit = []
for uid in range(num_units):
    spikes_for_uid = spike_times[unit_ids == uid].numpy()
    spikes_per_unit.append(spikes_for_uid)

plt.eventplot(spikes_per_unit)
plt.xlabel("time (s)")
plt.ylabel("Unit ID")
plt.title(f"Sample {sample_idx} spike-train")
plt.show()
```

## Loading model and pretrained weights

We'll now load the model, and apply pretrained weights to it.
As you can see in the printed logs, we were able to find the unit and session embeddings for this given
dataset in our checkpoint. This means our dataset was part of the pretraining of these weights.

```python
device = torch.device('cpu')

# Load model hyper-parameters from config
model_cfg = omegaconf.OmegaConf.load("../configs/model/poyo_single_session.yaml")
model_cfg.pop('_target_')

model = PerceiverNM(
    **model_cfg,
    unit_vocab=unit_vocab,
    session_names=session_names,
    use_memory_efficient_attn=False, # Since we're doing this on CPU
)

ckpt_path = "../logs/lightning_logs/f9sj5g0b/last.ckpt"
model.load_from_ckpt(ckpt_path)
model.to(device)
model.eval()
```

## Performing inference

Now, using this model, let's perform inference on our loaded data sample. Our model takes in spikes and outputs a 2D timeseries corresponding to the predicted hand-velocity.
We'll visualize our prediction relative to the ground truth. At the same time, let's also measure the accuracy of our inference in terms of its $R^2$ score.

```python
# Creating the collate function
collate_fn = Collate(
    num_latents_per_step=model_cfg.num_latents,
    step=1.0 / 8,
    sequence_length=1.0, # seconds
    unit_vocab=unit_vocab,
    metrics=[{"output_key": "CURSORVELOCITY2D"}],
)

# Test samples, in general, can be arbitrarily long in time.
# `stitched_prediction` does the job of breaking long samples into smaller chunks
# and then stitching together the results. 
# This is needed because POYO currently has a fixed context size of 1 second.
gt, pred = stitched_prediction(
    data=data_sample,
    collate_fn=collate_fn,
    model=model,
    device=device,
)
gt, pred = gt['CURSORVELOCITY2D'], pred['CURSORVELOCITY2D']

# Evaluating R2 score for our prediction
r2score = R2Score(num_outputs=gt.shape[1])
r2 = r2score(gt, pred).item()
print(f"R2 score: {r2:.3f}")

# Plotting the result
plt.plot(gt[:, 0], gt[:, 1], label="Ground truth")
plt.plot(pred[:, 0], pred[:, 1], label="Predicted")
plt.title(f"Sample {sample_idx}, $R^2$ = {r2:.3f}")
plt.legend()
plt.show()
```

## Evaluating accuracy over entire dataset

Now that we know how to perform inference and measure its accuracy over one sample, let's measure the average hand-velocity 

```python
gt_all, pred_all = [], []
for i, data_sample in enumerate(dataset):
    gt, pred = stitched_prediction(
        data=data_sample,
        collate_fn=collate_fn,
        model=model,
        device=device,
    )
    gt, pred = gt['CURSORVELOCITY2D'], pred['CURSORVELOCITY2D']

    gt_all.append(gt)
    pred_all.append(pred)

gt_all, pred_all = torch.cat(gt_all), torch.cat(pred_all)

r2 = r2score(gt_all, pred_all).item()
print(f"R2 score for entire dataset: {r2:.3f}")
```
